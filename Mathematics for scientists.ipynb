**2a) y = cosnx + n**

```python
# import libraries for plotting  
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
%matplotlib inline

# define functions (one containing both n, one with only the first n)
def f1(n,x):
    y = np.cos(n*x)
    return y 

def f2(n,x):
    y = np.cos(n*x)+ n
    return y 


# setting range for values 
n = np.linspace(-5,5,100)
x= np.linspace(-5,5,100)
X,Y =np.meshgrid(x,n)
y1 = f1(X,Y)
y2 = f2(X,Y)
```

```python
# make a plot to compare both functions 
fig = plt.figure(figsize= (10,4))
ax1 = plt.subplot(1,2,1)
ax1.contourf(y1, filled = True)
ax1.set_title('cos(nx)')
ax1.set_xlabel('x')
ax1.set_ylabel('n')
ax1.set_xticklabels(np.linspace(-5,5,6))
ax1.set_yticklabels(np.linspace(-5,5,6))


ax2 = plt.subplot(1,2,2)
ax2.contourf(y2, filled = True)
ax2.set_title('cos(nx) + n')
ax2.set_xlabel('x')
ax2.set_ylabel('n')
ax2.set_xticklabels(np.linspace(-5,5,6))
ax2.set_yticklabels(np.linspace(-5,5,6))

plt.show()
```

**Answer:** As the value of n changes, the effect of the first n is that the frequency of the cosinus wave changes, whereas the second n affects the amplitude of the wave. The two above figures show the solution of the two different cosinus functions for different n.

**3.) Plot function for various number ranges **

```python
# create ranges for different a
a1 = np.arange(-2,2,1)
a2 = np.arange(-3,3,0.5)
a3 = np.arange(-5,5,0.5)
print('a1', a1)
print('a2', a2)
print('a3',a3)
```

```python
def f(a1, a2, a3, x):
    y = a1/a3 * x - a2/a3
    return y 
```

```python
# make plot for a1, a2 and a3
x= np.arange(1,100)

for ai in a1:
    for aii in a2:
        for aiii in a3:
            if aiii != 0:
                plt.plot(x, f(ai,aii,aiii,x))
```

**4.) Plotting two functions side-by-side  ** 

```python
# define functions 

def f1(x):
    y = (x+2)**2 + 3/x
    return y 

def f2(x):
    y = -(2-x)**2 - 350 
    return y
    
x= np.arange(1,100)
y1= f1(x)
y2= f2(x)
```

```python
# create side-by-side plot 
plt.figure(figsize=(10,4))

ax1= plt.subplot(1,2,1)
ax1.plot(x,y1, color= 'black')

ax2= plt.subplot(1,2,2)
ax2.plot(x,y2, color = 'blue')

plt.show()
```

**5.) Surface and Density plot **

```python
# define function 
def f(x,y):
    z = x*y + y
    return z 

# setting range for values 
x = np.linspace(-10,10,100)
y= np.linspace(-10,10,100)
X,Y= np.meshgrid(x,y)
z= f(X,Y)
```

```python
plt.figure(figsize=(20,8))

ax1= plt.subplot(1,2,1, projection='3d')

# get checkerboard pattern for 3D surface plot
colortuple = ('y', 'b')
xlen= len(x)
ylen= len(y)
colors = np.empty(X.shape, dtype=str)
for y in range(ylen):
    for x in range(xlen):
        colors[x, y] = colortuple[(x + y) % len(colortuple)]
        
ax1.plot_surface(X, Y, z, rstride=1, cstride=1,facecolors= colors, edgecolor='none')
ax1.set_title('surface plot')

ax2= plt.subplot(1,2,2)
m = ax2.pcolormesh(X,Y,z, cmap= 'Greys')
ax2.set_title('Density plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.colorbar(m)
plt.show()
```

**6.) Polynomial** 


```python
lambda_1 = 4 
lambda_2 = 3 
lambda_3 = -5 

# import equation solver 
from sympy import * 
x = Symbol("x")
y = Symbol("y")  
equation = Eq(lambda_1 * x**2 + lambda_2 * x + lambda_3, 0)
print('solution for polynom:', solve(equation, x))
```

```python
# plot function 

def f(x):
    y = lambda_1 * x**2 + lambda_2 * x + lambda_3
    return y 

x= np.arange(-10,10)
y= f(x)
plt.plot(x,y)
plt.show()
```

**7.) Depth profile**

```python
fig = plt.figure(figsize= (10,4))

depth = [0.5, 1, 2, 4, 12]
concentration = [52, 23, 17, 14, 9]


ax1 = plt.subplot(1,2,1)
ax1.plot(concentration, depth)
ax1.set_ylabel('depth')
ax1.set_xlabel('concentration')
# reverse y axis to show depth profile
ax1.invert_yaxis()


ax2 = plt.subplot(1,2,2)
ax2.plot(np.log(depth), np.log(concentration))
logmean = np.mean(np.log(concentration))
ax2.plot(logmean, marker = 'o', markersize= 10, label= 'log mean')
ax2.set_ylabel('log depth')
ax2.set_xlabel('log concentration')
ax2.invert_yaxis()
plt.legend()
plt.show()
```
